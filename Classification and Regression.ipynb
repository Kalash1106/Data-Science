```python
split_ratio=[0.8,0.6,0.7,0.9] #Ratios to be used for the question
for n in split_ratio:
    print('\033[1m' + 'Linear Regression analysis for split ratio ' + str(int(n*100)) +":"+ str(int(100- n*100)) + '\033[0m')
    n_rows=np.ceil(n*413) #number of rows to be included in the train data

    train_data=df1.iloc[0:int(n_rows)]      #Defining the test and train data
    test_data=pd.DataFrame(df1.iloc[int(n_rows):df1.shape[0]])
   
    
    LR.fit(train_data.drop(columns=['Y house price of unit area']),  train_data.iloc[0:train_data.shape[0],6]) #fitting the data
    
    print("Intercept of model is", LR.intercept_)      #.intercept_ prints the intercept
    
    display(pd.DataFrame(LR.coef_,index=[1,2,3,4,5,6]).rename(columns={0:'Weights'}))  # .coef_ gives the coefficients
    print(" ")

    y_prediction=LR.predict(test_data.drop(columns=['Y house price of unit area']))  #.predict(test_data) helps to predict the data
    y_prediction_data=pd.DataFrame(y_prediction).rename(columns={0:'Predicted Prices'})
    print(" The predicted data is:")
    display(y_prediction_data)
    print(" ")

    from sklearn.metrics import mean_squared_error   #Mean Squared error
    print("Mean squared error is" , mean_squared_error(test_data.iloc[0:test_data.shape[0],6],y_prediction))
    print(" ")

    print('Coefficient of determination:', LR.score(train_data.drop(columns=['Y house price of unit area']),train_data.iloc[0:train_data.shape[0],6])) #calculating r^2 using .score(x,y)
    print(" ")
    print(" ")



    
   
```

```python
split_ratio=[0.8,0.6,0.7,0.9] #Ratios to be used for the question
for n in split_ratio:
    print('\033[1m' + 'Ridge Regression analysis for split ratio ' + str(int(n*100)) +":"+ str(int(100- n*100)) + '\033[0m')
    n_rows=np.ceil(n*413) #number of rows to be included in the train data

    train_data=df1.iloc[0:int(n_rows)]      #Defining the test and train data
    test_data=pd.DataFrame(df1.iloc[int(n_rows):df1.shape[0]]) 
    # Train the model
    lambda_values=[.001, 0.005, 0.01, 0.05, 0.1,0.5]
    for l in lambda_values:
        ridgeR = Ridge(alpha = l)
        ridgeR.fit(train_data.drop(columns=['Y house price of unit area']),  train_data.iloc[0:train_data.shape[0],6])
        y_pred = ridgeR.predict(test_data.drop(columns=['Y house price of unit area']))
        print("For regularisation parameter (alpha)=", l)
    
 
    # calculate mean square error
        mean_squared_error_ridge = np.mean((y_pred - test_data.iloc[0:test_data.shape[0],6])**2)
        print("Mean Squared Error is" , mean_squared_error_ridge)
    
 
     # get r^2 value for Ridge
        print("r^2 value is:",r2_score(test_data.iloc[0:test_data.shape[0],6], y_pred))
        print("")
        print("")
      

```

Q2 2] Loading the dataset and doing pre-processing


```python
df2=pd.read_csv('/content/haberman.csv', header=None)
df2=df2.rename(columns={0:'Age', 1:'Year of operation',2:'Number of nodes',3:'Survival Status'})
display(df2)
```

Q2 3,4,5] Splitting the dataset

```python
n_split=[0.8,0.6,0.7,0.9]  #Splitting ratios
tot_rows=df2.shape[0] #Number of rows in dataset 2
for n in n_split:
    print('\033[1m' + 'Logistic Regression analysis for split ratio ' + str(int(n*100)) +":"+ str(int(100- n*100)) + '\033[0m')
    n_rows=int(np.ceil(n*tot_rows))  #n_rows will be considered for train data, the rest into test data

    x_train=df2.iloc[0:n_rows, 0:3] #data matrix for train data
    y_train=df2.iloc[0:n_rows, 3]  #y column for train data
    x_test= df2.iloc[n_rows:tot_rows, 0:3] #data matrix for testing
    y_test=pd.DataFrame(df2.iloc[n_rows:tot_rows, 3])   #y column for testing

    #Finding coefficents and intercepts on the model trained on logistic regression
    model.fit(x_train,y_train)
    print("Intercept of the model is", model.intercept_)
    print("The coefficients are", model.coef_)


    #Predicting the Survival status 
    y_pred=model.predict(x_test)
    display(pd.DataFrame(y_pred).rename(columns={0:'Prediction'}))
    
    #Cumulative Classification error from loops
    cm_error=0
    c_error=0
    n_obs=int(np.ceil(y_pred.shape[0]*1))
    for i in range(0,n_obs,1):
        cm_error=cm_error + np.abs(int(pd.DataFrame(y_pred).iloc[i,0]-int(y_test.iloc[i,0])))
    c_error=cm_error/y_test.shape[0]
    print("The classification error is" ,c_error) 
    print(" ")
    print(" ")
```
